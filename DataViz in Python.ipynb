```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Basic Graph

```python
x=[0,1,2,3,4,5]
y=[0,2,4,6,8,10]
plt.plot(x,y,label='2x',color='red',linewidth=2,marker='.',markersize=10,markeredgecolor='blue')

#line 2
x2=np.arange(0,4.5,0.5)
#[0,0.5,1,1.5,2,2.5,3,3.5,4,4.5]
plt.plot(x2,x2**2,"b",label='x^2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("line graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})

#lt.xticks([0,1,2,3,4,5])
#t.yticks([0,2,4,6,8,10])

#shorthand: '[color][marker][line]' : r.-

#adding legend
plt.legend()


plt.show()
```

```python
#line 2
x2=np.arange(0,4.5,0.5)
#[0,0.5,1,1.5,2,2.5,3,3.5,4,4.5]
#resize graph
plt.figure(figsize=(5,3),dpi=300)
plt.plot(x2[:4],x2[:4]**2)
plt.plot(x2[3:],x2[3:]**2,'r--')
#show projection[:4]^^

#save graph
#plt.savefig('mygraph.png',dpi=300)
plt.show()
```

```python
#Bar Charts
labels=['a','b','c']
values=[1,4,3]
bars=plt.bar(labels,values)
plt.figure(figsize=(6,4))
#adding patterns
patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.show()
```
